```python
# Multiple Linear Regression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

url = "https://raw.githubusercontent.com/rajendra0968jangid/ml-project-1/main/student_performance.csv"
df = pd.read_csv(url)

X = df[['Study_Hours','Attendance','Internal_Marks','Assignment_Marks','Practical_Marks']]

Y = df[['Final_Score']]

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

model = LinearRegression()

model.fit(X_train,Y_train)

prediction = model.predict(X_test)

print("Multiple Linear Regression Actual:")
print(Y_test.values)

print("Multiple Linear Regression Prediction:")
print(prediction)
```
